Este archivo lo que hace es generar información rápida sobre toda las TF de la simulación.

Se grafican las TF de las variables de una simulación en un punto.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy.polynomial.polynomial as poly
import matplotlib
import matplotlib.patches as mpatches
```

```python
pathrw = '/home/hcapettini/Escritorio/HDD1TB/Tesis/Analísis/Mayo/'
#pathrw = '/media/adminpc/Hilario-2/Tesis/Analísis/Mayo/'


simulacion='pulso_4' 
#simulacion='desequilibrio_5' 

pos= '0.2_20.0'
tint='100_500_'
fmin = 0.
fmax = 0.02

path=pathrw +simulacion+'/'+simulacion+'_'+pos+'_TF_'+tint
```

```python
with open( path + 'velx.txt') as velx:
    velxlines = velx.readlines()[2:]
    time  = [float(line.split()[0]) for line in velxlines]
    velxx = [float(line.split()[1]) for line in velxlines]
with open( path + 'vely.txt') as vely:
    velylines = vely.readlines()[2:]
    velyy = [float(line.split()[1]) for line in velylines]  
with open( path + 'magx.txt') as magx:
    magxlines = magx.readlines()[2:]
    magxx = [float(line.split()[1]) for line in magxlines]
with open( path + 'magy.txt') as magy:
    magylines = magy.readlines()[2:]
    magyy = [float(line.split()[1]) for line in magylines]
with open( path + 'dens.txt') as dens:
    denslines = dens.readlines()[2:]
    densy = [float(line.split()[1]) for line in denslines]
with open( path + 'pres.txt') as pres:
    preslines = pres.readlines()[2:]
    presy = [float(line.split()[1]) for line in preslines]
```

Acá de determina la densidad de energía del punto

```python
velx = np.array(velxx)
vely = np.array(velyy)
magx = np.array(magxx)
magy = np.array(magyy)
dens = np.array(densy)
pres = np.array(presy)
```

```python
plt.plot(time,velx,label='Vr en'+''+'(r,z)='+pos)
plt.legend()
plt.xlim(fmin, fmax)
#plt.xlim(0.07, 0.08)
plt.ylim(0, 35.e5)
plt.ticklabel_format(axis="y", style="sci", scilimits=(5,5))
plt.tick_params( labelsize=17)
plt.minorticks_on()
plt.ylabel('Magnitud de Fourier Vr [cm/s]',fontsize=17) 
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(path+'vz.pdf', bbox_inches='tight')   
```

```python
plt.plot(time,vely,label='Vz en'+''+'(r,z)='+pos)
plt.legend()
plt.xlim(fmin, fmax)
#plt.xlim(0.03, 0.08)

plt.ylim(0.e5, 13000.e5)
plt.ticklabel_format(axis="y", style="sci", scilimits=(5,5))
plt.tick_params( labelsize=17)
plt.minorticks_on()
plt.ylabel('Magnitud de Fourier Vz [cm/s]',fontsize=17) 
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(path +'vz.pdf', bbox_inches='tight')   
```

```python
plt.plot(time,magx,label='Br en'+''+'(r,z)='+pos)
plt.legend()
plt.xlim(fmin, fmax)
#plt.xlim(0., 0.02)
#plt.ylim(0, 3.)
plt.tick_params( labelsize=17)
plt.minorticks_on()
plt.ylabel('Magnitud de Fourier Br [G]',fontsize=17) 
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(path+'_Br.pdf', bbox_inches='tight')   
```

```python
plt.plot(time,magy,label='Bz en'+''+'(r,z)='+pos)
plt.legend()
plt.xlim(fmin, fmax)
#plt.xlim(0., 0.02)
plt.ylim(0, 280)
plt.tick_params( labelsize=17)
plt.minorticks_on()
plt.ylabel('Magnitud de Fourier Bz [G]',fontsize=17) 
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(path+'_Bz.pdf', bbox_inches='tight')   
```

```python
plt.plot(time,dens,label='Dens en'+''+'(r,z)='+pos)
plt.legend()
plt.xlim(fmin, fmax)
#plt.xlim(0.15, 0.17)
plt.ylim(0., 90.e-14)
plt.ticklabel_format(axis="y", style="sci", scilimits=(-14,-14))
plt.tick_params( labelsize=17)
plt.minorticks_on()
plt.ylabel('Magnitud de Fourier densidad [gr/cm^3]',fontsize=17) 
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(path+'_dens.pdf', bbox_inches='tight')   
```

```python
plt.plot(time,pres,label='Pres en'+''+'(r,z)='+pos)
plt.legend()
plt.xlim(fmin, fmax)
#plt.xlim(0.005, 0.015)
plt.ylim(0., 1050)
plt.tick_params( labelsize=17)
plt.minorticks_on()
plt.ylabel('Magnitud de Fourier presión [dyn/cm^2]',fontsize=17)
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(path+'_pres.pdf', bbox_inches='tight')   
```

Comparación de variables normalizadas

```python
velxn = (velx)/np.max(velx)
velyn = (vely)/np.max(vely)
magxn = (magx)/np.max(magx)
magyn = (magy)/np.max(magy)
densn = (dens)/np.max(dens)
presn = (pres)/np.max(pres)
```

```python
vr,  = plt.plot(time,velxn,label='Vr')
vz,  = plt.plot(time,velyn,label='Vz')
Br,  = plt.plot(time,magxn,label='Br')
Bz,  = plt.plot(time,magyn,label='Bz')
Dens,= plt.plot(time,densn,label='Densidad')
Pres,= plt.plot(time,presn,label='Presión')

plt.legend(handles=[vr,vz, Br, Bz,Dens,Pres])

plt.xlim(fmin, fmax)
#plt.xlim(0, 0.1)

#plt.ylim(emin, emax)
plt.title('Equilibrio')
plt.xlabel('Frecuencia [Hz]',fontsize=17)
#plt.savefig(pathrw+simulacion +'/'+pos+'_variables_normalizadas.pdf', bbox_inches='tight')   
```
