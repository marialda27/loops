Se grafican la evolución de las variables de una simulación en un punto
en conjunto con la evolución de la densidad de energía.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy.polynomial.polynomial as poly
import matplotlib
import matplotlib.patches as mpatches
```

```python
path = '/home/hcapettini/Escritorio/HDD1TB/Tesis/Analísis/Mayo/'
#path = '/media/adminpc/Hilario-2/Tesis/Analísis/Mayo/'

#simulacion ='pulso_8'
simulacion ='desequilibrio_3'

pos= '0.2_20.0'
#pos= '1.5_20.0'
#pos= '2.8_20.0'
#pos= '0.2_29.7'
#pos= '1.5_29.7'
#pos= '2.8_29.7'
#pos= '0.2_23.0'

tmin = 0
tmax = 1000

pathr=path +simulacion+'/'+simulacion+'_'+pos+'_'
```

```python
with open( pathr + 'velx.txt') as velx:
    velxlines = velx.readlines()[2:]
    time  = [float(line.split()[0]) for line in velxlines]
    velxx = [float(line.split()[1]) for line in velxlines]
with open( pathr + 'vely.txt') as vely:
    velylines = vely.readlines()[2:]
    velyy = [float(line.split()[1]) for line in velylines]  
with open( pathr + 'magx.txt') as magx:
    magxlines = magx.readlines()[2:]
    magxx = [float(line.split()[1]) for line in magxlines]
with open( pathr + 'magy.txt') as magy:
    magylines = magy.readlines()[2:]
    magyy = [float(line.split()[1]) for line in magylines]
with open( pathr + 'dens.txt') as dens:
    denslines = dens.readlines()[2:]
    densy = [float(line.split()[1]) for line in denslines]
with open( pathr + 'pres.txt') as pres:
    preslines = pres.readlines()[2:]
    presy = [float(line.split()[1]) for line in preslines]
```

Acá de determina la densidad de energía del punto

```python
velx = np.array(velxx)
vely = np.array(velyy)
magx = np.array(magxx)
magy = np.array(magyy)
dens = np.array(densy)
pres = np.array(presy)
```

```python
gama = 5/3
```

```python
ek = 1/2 * dens *(velx**2+vely**2)
ei = pres /(gama-1)
em = (magx**2+magy**2)/(8*np.pi)
```

```python
e = ek + ei + em
```

```python
emax = 300
emin = -1
```

```python
cinetico,  = plt.plot(time,ek,label='Cinética')
magnetico, = plt.plot(time,em,label='Magnética')
interno,   = plt.plot(time,ei,label='Interna')
total,     = plt.plot(time,e,label='Total')
plt.legend(handles=[cinetico,magnetico, interno, total])
plt.tick_params( labelsize=17)
plt.xlim(tmin, tmax)
#plt.ylim(emin, emax)
#plt.title('Evolución de las densidades de energía',fontsize=17)
plt.ylabel('Densidad de energía',fontsize=17)
plt.xlabel('Tiempo',fontsize=17)
#plt.savefig(path+simulacion +'/'+pos+'_energia.pdf', bbox_inches='tight')   
```

Comparación de variables normalizadas

```python
velxn = (velx-velx[0])/np.max(velx)
velyn = (vely-vely[0])/np.max(vely)
magxn = (magx-magx[0])/np.max(magx)
magyn = (magy-magy[0])/np.max(magy)
densn = (dens-dens[0])/np.max(dens)
presn = (pres-pres[0])/np.max(pres)
```

```python
vr,  = plt.plot(time,velxn,label='Vr')
vz,  = plt.plot(time,velyn,label='Vz')
Br,  = plt.plot(time,magxn,label='Br')
Bz,  = plt.plot(time,magyn,label='Bz')
Dens,= plt.plot(time,densn,label='Densidad')
Pres,= plt.plot(time,presn,label='Presión')

plt.legend(handles=[vr,vz, Br, Bz,Dens,Pres])

plt.xlim(tmin, tmax)
#plt.ylim(emin, emax)
plt.title('Equilibrio')
plt.xlabel('Tiempo')
#plt.savefig(path+simulacion +'/'+pos+'_variables_normalizadas.pdf', bbox_inches='tight')   
```
