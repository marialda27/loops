```python
#pathrw='/home/hcapettini/Escritorio/HDD1TB/Tesis/Analísis/Mayo/'
pathrw = '/media/usuario/Hilario-2/Tesis/Analísis/Mayo/'

#simulacion ='pulso_3'
simulacion ='desequilibrio_3'

pos1= '0.2_20.0'
tint='0_500_'
fmin = 0.
fmax =40.
path=pathrw +simulacion+'/'+simulacion+'_'+pos1+'_TF_'+tint
#path = path+ pos1+'_TF_'+tint
```

```python
with open( path + 'energy.txt') as energy:
    energylines = energy.readlines()[0:]
    time  = [float(line.split()[0]) for line in energylines]
    energyx = [float(line.split()[1]) for line in energylines]
with open( path + 'energy_i.txt') as energy_i:
    energy_ilines = energy_i.readlines()[0:]
    energy_iy = [float(line.split()[1]) for line in energy_ilines]
    
with open( path + 'energy_m.txt') as energy_m:
    energy_mlines = energy_m.readlines()[0:]
    energy_my = [float(line.split()[1]) for line in energy_mlines]
    
with open( path + 'energy_k.txt') as energy_k:
    energy_klines = energy_k.readlines()[0:]
    energy_ky = [float(line.split()[1]) for line in energy_klines]
    
```

```python
energy,  = plt.plot(time,energyx,label='E')
energy_i,  = plt.plot(time,energy_iy,label='E_i')
energy_m,  = plt.plot(time,energy_my,label='E_m')
energy_k,  = plt.plot(time,energy_ky,label='E_k')

plt.legend(handles=[energy,energy_i,energy_m,energy_k])

plt.xlim(0.0, 0.2)
#plt.ylim(0, 4500)
plt.title('Caso II' +'en '+pos1)
plt.xlabel('Frequencia [Hz]')
plt.ylabel('Magnitud de Fourier')
#plt.savefig(pathrw+simulacion+ '/'+simulacion+'_'+'TF.pdf', bbox_inches='tight') 
```
